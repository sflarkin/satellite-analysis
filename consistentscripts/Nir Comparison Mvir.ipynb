```python
import glob
import numpy as np
import matplotlib.pyplot as plt
from satellite_analysis.catalogreaders import consistentcatalogreader as consistent
from satellite_analysis.catalogreaders import tomercatalogreader as tomer
from yt.utilities.cosmology import Cosmology


#first find all of the VELA dirs to iterate over
input_dir = '/Users/user1/documents/GEN3/rockstarcomoving001'
output_dir ='/Users/user1/documents/GEN3/rockstarcomoving001/NirMvirComparison'

#extract the VELA number for each sim to feed to tomer's catalog later
catalog_files = glob.glob('{}/VELA*'.format(input_dir))
#catalog_files = ['/Users/user1/documents/GEN3/VELA15/rockstar001']
catalog_numbers = []
for file in catalog_files:
    catalog_numbers.append(file[-2:])
catalog_numbers = np.unique(catalog_numbers)
catalog_numbers.sort()
#catalog_numbers = ['15']
#now loop over each VELA sim to make the plot for each

co = Cosmology(hubble_constant = .70, omega_matter = .27, omega_lambda = .73)

for index in range(len(catalog_files)):
    #load the consistent catalog
    consistent.consistent_catalog_reader('{}/*hlists'.format(catalog_files[index]), remove_subhalos='True')

    #load the tomer catalogs
    tomer.read_tomer(catalog_numbers[index])
    tomer_scales = [str(scale)[2:5] for scale in tomer.tomer_list['aexpn'].tolist()]
    r_vir_tomer = tomer.tomer_list['r_vir[kpc]'].tolist()
    m_vir_tomer = tomer.tomer_list['M_rvir[Msun]'].tolist()
    
    #create the lists for the data to be stored in
    rockstar_scales = []
    rockstar_rvirs = []
    rockstar_mvirs = []
    rockstar_deltas = []
    
    rockstar_comparison_rvir = []
    tomer_comparison_rvir = []
    rockstar_comparison_mvir = []
    tomer_comparison_mvir = []
    rockstar_comparison_delta = []
    tomer_comparison_delta = []
    #now loop over each snapshot in each simulation
    for scale_index in consistent.snapshot_index:
        #find the scale factor and rvir(converting from comoving Mpc/h to proper Mpc) for the consistent catalogs
        VELA_a = consistent.consistent_file_index[scale_index]
        a = float(VELA_a)/1000
        hz = .70*(((.27*((1/a)**3)) + .73)**(1/2))
        #for the matter density and lambda density, do the rho of the rho/rho(c) stay the same, and that is how
        #I can find how they change with time?
        rvir = float(consistent.halo_data_sorted[scale_index][0][11]) * a /.7
        mvir = float(consistent.halo_data_sorted[scale_index][0][10]) /.7
        critical_density = co.critical_density((1/a) -1)
        critical_density_0 = co.critical_density(0)
        delta_vir_floats = (mvir/((4/3)*np.pi*(rvir*rvir*rvir)))/((float(critical_density.in_units('Msun*h**2/kpc**3')))*(.27/(a**3)))

        rockstar_scales.append(VELA_a)
        rockstar_rvirs.append(rvir)
        rockstar_mvirs.append(mvir/10**12)
        rockstar_deltas.append(delta_vir_floats /(.7**2))
        #rvir_times.append(rvirtimes)
        
        #see if there is a scale factor that matches for the direct comparison plot between the two catalogs
        tomer_number = [pos for pos, number in enumerate(tomer_scales) if number == VELA_a]
        #if there is a match in scale factor, find the rvir
        if tomer_number != []:
            tomer_rvir = float(r_vir_tomer[tomer_number[0]])
            tomer_mvir = float(m_vir_tomer[tomer_number[0]])
            tomer_delta_vir_floats = (tomer_mvir/((4/3)*np.pi*(tomer_rvir**3)))/(float(critical_density.in_units('Msun/kpc**3')))/critical_density_0
            tomer_comparison_rvir.append(tomer_rvir)
            rockstar_comparison_rvir.append(rvir)
            tomer_comparison_mvir.append(tomer_mvir/10**12)
            rockstar_comparison_mvir.append(mvir/10**12)
            tomer_comparison_delta.append(tomer_delta_vir_floats)
            rockstar_comparison_delta.append(delta_vir_floats/(.7**2))
        
    #convert the scales to redshifts for plotting
    Nir_redshifts = [(1/(float(x)/1000)-1) for x in tomer_scales]
    rockstar_redshifts = [(1/(float(x)/1000)-1) for x in rockstar_scales]
    
    #convert the strings to floats for tomer_scales
    r_vir_tomer = [float(x) for x in r_vir_tomer]
    m_vir_tomer = [float(x)/10**12 for x in m_vir_tomer]
    tomer_deltas = [((m_vir_tomer[x]*10**12)/((4/3)*np.pi*(r_vir_tomer[x]**3)))/(float(co.critical_density(Nir_redshifts[x]).in_units('Msun/kpc**3')))/critical_density_0 for x in range(len(m_vir_tomer))]
    #now plot the two graphs. The one on the left is a direct comparsion, while the one on the right is each
    #sims individual evolution according to each catalog
    
    
    plt.figure(1, figsize=(20,30))
    plt.subplot(321)
    plt.title('VELA %s' % catalog_numbers[index])
    plt.xlabel('Rockstar Rvir (Kpc)')
    plt.ylabel('Nir Rvir (Kpc)')
    #plt.xscale('log')
    #plt.yscale('log')
    largest_rvir = np.amax([np.amax(rockstar_comparison_rvir), np.amax(tomer_comparison_rvir)]) + 10
    plt.xlim(0, largest_rvir)
    plt.ylim(0, largest_rvir)
    plt.scatter(rockstar_comparison_rvir, tomer_comparison_rvir, s = 5)
    x = np.linspace(0, largest_rvir, largest_rvir)
    plt.plot(x, x, '--m')
    
    plt.subplot(322)
    plt.xlabel('Redshift')
    plt.ylabel('Virial Radius (Kpc)')
    #plt.yscale('log')
    #plt.ylim(10**(7.8), 10**(14))
    plt.scatter(Nir_redshifts, r_vir_tomer, s = 5)
    plt.scatter(rockstar_redshifts, rockstar_rvirs, s = 5, c = 'r')
    #plt.scatter(rockstar_redshifts, rvir_times, s = 5)
    plt.legend(('Nir Catalog', 'Rockstar Catalog'))
    
    plt.subplot(323)
    #plt.title('VELA %s' % catalog_numbers[index])
    plt.xlabel('Rockstar Mvir (10^12 MSun)')
    plt.ylabel('Nir Mvir (10 ^12 Msun)')
    #plt.xscale('log')
    #plt.yscale('log')
    largest_mvir = np.amax([np.amax(rockstar_comparison_mvir), np.amax(tomer_comparison_mvir)]) + .1
    smallest_mvir = np.amin([np.amin(rockstar_comparison_mvir), np.amin(tomer_comparison_mvir)]) - .1
    plt.xlim(smallest_mvir, largest_mvir)
    plt.ylim(smallest_mvir, largest_mvir)
    plt.scatter(rockstar_comparison_mvir, tomer_comparison_mvir, s = 5)
    x = np.linspace(-.5, 2, 50)
    plt.plot(x, x, '--m')
    
    plt.subplot(324)
    plt.xlabel('Redshift')
    plt.ylabel('Virial Mass (10^12 Msun)')
    plt.yscale('log')
    #plt.ylim(10**(7.8), 10**(14))
    plt.scatter(Nir_redshifts, m_vir_tomer, s = 5)
    plt.scatter(rockstar_redshifts, rockstar_mvirs, s = 5, c = 'r')
    #plt.scatter(rockstar_redshifts, rvir_times, s = 5)
    plt.legend(('Nir Catalog', 'Rockstar Catalog'))
    
    plt.subplot(325)
    plt.title('VELA %s' % catalog_numbers[index])
    plt.xlabel('Rockstar Delta Vir (Kpc)')
    plt.ylabel('Nir Delta Vir (Kpc)')
    #plt.xscale('log')
    #plt.yscale('log')
    smallest_delta = np.amin([np.amin(rockstar_comparison_delta), np.amin(tomer_comparison_delta)]) - 10
    largest_delta = np.amax([np.amax(rockstar_comparison_delta), np.amax(tomer_comparison_delta)]) + 10
    plt.xlim(smallest_delta, largest_delta)
    plt.ylim(smallest_delta, largest_delta)
    plt.scatter(rockstar_comparison_delta, tomer_comparison_delta, s = 5)
    x = np.linspace(smallest_delta, largest_delta, 100)
    plt.plot(x, x, '--m')
    
    plt.subplot(326)
    plt.xlabel('Redshift')
    plt.ylabel('Delta Vir')
    #plt.yscale('log')
    #plt.ylim(10**(7.8), 10**(14))
    plt.scatter(Nir_redshifts, tomer_deltas, s = 5)
    plt.scatter(rockstar_redshifts, rockstar_deltas, s = 5, c = 'r')
    #plt.scatter(rockstar_redshifts, rvir_times, s = 5)
    plt.legend(('Nir Catalog', 'Rockstar Catalog'))
    plt.savefig('%s/VELA%sdeltavirdividetests.7.png' % (output_dir, catalog_numbers[index]))
    plt.close()
```
