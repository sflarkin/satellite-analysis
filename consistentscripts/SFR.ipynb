```python
from satellite_analysis.catalogreaders import consistentcatalogreader as consistent
import glob

#input_dirs = glob.glob('/Users/user1/documents/GEN6/rockstarcomoving001/VELA*/hlists')

input_dirs = glob.glob('/Users/user1/documents/GEN3/rockstar_moresnaps/VELA*/hlists')

projenitor_information = []

for input_dir in input_dirs:
    consistent.consistent_catalog_reader(input_dir, remove_subhalos='False')
    VELA_number = input_dir[-9:-7]
    print(VELA_number)

    count = len(consistent.consistent_file_index) - 1

    while count > 0:
        scale = consistent.consistent_file_index[count]
        #print(scale)
        projenitor_scale = consistent.consistent_file_index[count-1]
    
        snap_to_find_projenitors = consistent.halo_data_sorted[count]
    
        snap_of_projenitors = consistent.halo_data_sorted[count - 1]
    
    
    
        for halos_to_find_projenitors in snap_to_find_projenitors:
            halo_id = halos_to_find_projenitors[1]
        
            largest_mass = 0
            projenitor_id = None 
        
            for halos_that_could_be_projenitors in snap_of_projenitors:
                if halos_that_could_be_projenitors[3] == halo_id and float(halos_that_could_be_projenitors[10]) > largest_mass:
                    largest_mass = float(halos_that_could_be_projenitors[10])
                    projenitor_id = halos_that_could_be_projenitors[1]
            if projenitor_id != None:
                projenitor_information.append((VELA_number, scale, halo_id, projenitor_scale, projenitor_id))
        count = count - 1
             
```

```python
#do a double loop over all fo the data to find the matching halos
#need to find both the projenitor and same halo

from satellite_analysis.consistentscripts import GEN3times as times
final_storage = []

scales = np.unique([x[1] for x in projenitor_information])
scales_remainder = np.unique([x[1] for x in projenitor_information if (int(x[1]) % 10 == 0)])


print(scales)

for scale in scales:
    if scale not in scales_remainder:
        continue
    print('Generating graph for scale {}'.format(scale))
    VELA_numbers = np.unique([x[0] for x in projenitor_information if x[1]==scale])
    
    gas_mass_list, stellar_mass_list, baryon_mass_list = [], [], []
    gas_ratio_list, stellar_ratio_list, baryon_ratio_list = [], [], []
    
    halo_mass_list = []
    SFR_list, SSFR_list = [], []
    
    stellar_mass_change_list = []
    ellapsed_time_list = []
    
    
    #now that we have all the VELA's that have that scale, find all of the halos of each combo of VELA and scale 
    #to find the matching tuples for graph generation
    
    for VELA_number in VELA_numbers:
        #print('Generating SFR for VELA {}'.format(VELA_number))
        projenitor_data = [x for x in projenitor_information if x[0] == VELA_number and x[1] == scale]
        
        
        #get the data for the baryons of the same snap and the previous snap for the projenitor
        baryon_data_current_halo = [x for x in baryon_information if x[0] == VELA_number and x[1] == projenitor_data[0][1]]
        baryon_data_projenitor = [x for x in baryon_information if x[0] == VELA_number and  x[1] == projenitor_data[0][3]]
        
        times_current_halo = [x for x in times.times_list if x[0] == VELA_number and x[1] == projenitor_data[0][1]]
        times_projenitor = [x for x in times.times_list if x[0] == VELA_number and x[1] == projenitor_data[0][3]]
        print(projenitor_data[0][1], projenitor_data[0][3])
        if times_projenitor == [] or times_current_halo == []:
            continue 
        #print(times_current_halo, times_projenitor)
        ellapsed_time = (times_current_halo[0][2] - times_projenitor[0][2]) * 10**9
        
        #now find the halos with the correct ids to match the current halo and projenitor
        current_halo = False
        projenitor = False
        for halo in projenitor_data:
            
            for current_halo_candidates in baryon_data_current_halo:
                if halo[2] == current_halo_candidates[2][0]:
                    current_halo = current_halo_candidates
                    
            for projenitor_halo_candidates in baryon_data_projenitor:   
                if halo[4] == projenitor_halo_candidates[2][0]:
                    projenitor = projenitor_halo_candidates
            
            if current_halo != False and projenitor !=False:
                #if both the projenitor and current halo data exist for the stars, find the values to plot and
                #save them to the lists
                
                #need gas mass, stellar mass, Mvir, SFR baryon mass 
                gas_mass = float(current_halo[2][8])
                #stellar_mass = float(current_halo[2][7])
                #stellar_mass_010 = float(current_halo[2][4])
                #stellar_mass_015 = float(current_halo[2][5])
                stellar_mass_020 = float(current_halo[2][6])
                dark_matter_mass = float(current_halo[2][9])
                
                #which dark_matter_mass should I use
                
                #should I use 
                
                baryon_mass = gas_mass + stellar_mass_020
                halo_mass = baryon_mass + dark_matter_mass
                
                gas_ratio = gas_mass / halo_mass
                stellar_ratio = stellar_mass_020 / halo_mass
                baryon_ratio = baryon_mass / halo_mass
                
                #now get the stellar data for the projenitor for the calculation of SFR
                
                #projenitor_stellar_mass = float(projenitor[2][7])
                #projenitor_stellar_mass_010 = float(projenitor[2][4])
                #projenitor_stellar_mass_015 = float(projenitor[2][5])
                projenitor_stellar_mass_020 = float(projenitor[2][6])
                
                #what units should these be in?
                stellar_mass_change = stellar_mass_020 - projenitor_stellar_mass_020
                stellar_mass_change_list.append(stellar_mass_change)
                SFR = (stellar_mass_020 - projenitor_stellar_mass_020) / ellapsed_time
                if stellar_mass_020 == 0:
                    SSFR = 0
                else:
                    SSFR = SFR / stellar_mass_020
                
                #now add the values to the lists for plotting once all of the VELAS of a specific scale_factor
                #have been looped through
                
                gas_mass_list.append(gas_mass)
                gas_ratio_list.append(gas_ratio)
                stellar_mass_list.append(stellar_mass_020)
                stellar_ratio_list.append(stellar_ratio)
                baryon_mass_list.append(baryon_mass)
                baryon_ratio_list.append(baryon_ratio)
                halo_mass_list.append(halo_mass)
                
                SFR_list.append(SFR)
                SSFR_list.append(SSFR)
                
    #now plot the data
    x = np.linspace(10**7.8, 10**14, 3000)
    y1 = [.167] * 3000
    y2 = [q*.167 for q in x]
    
    #print(ellapsed_time_list)
    a = float(scale)
    plt.figure(1, figsize=(20,20))
    
    #plt.subplot(521)
    #plt.title('Scale Factor %s, Gas Mass Ratio vs Halo Mass' % str(a/1000))
    #plt.xlabel('Halo Mass (Gas Mass @ Rvir + Dark Matter Mass @ Rvir + Stellar Mass @ .2Rvir)')
    #plt.ylabel('Gas Mass / Halo Mass')
    #plt.xscale('log')
    #plt.yscale('log')
    #plt.xlim(10**(7.8), 10**(14))
    ##plt.ylim(0.0, 0.10)
    #plt.plot(x, y1)
    #plt.scatter(halo_mass_list, gas_ratio_list, c=SFR_list, s = 1, cmap='RdBu')
    #cbar= plt.colorbar()
    #cbar.set_label('SFR in MSun/yr')
    #plt.colorbar.set_label('SFR in MSun/yr')
    #plt.legend(('Omega Baryon / Omega Matter = 0.167',  'Host Galaxies'))
    
    plt.subplot(221)
    plt.title('Gas Mass vs Halo Mass at Redshift %s' % str((1/(a/1000)) - 1)[0:3])
    plt.xlabel('Halo Mass (Gas Mass @ Rvir + Dark Matter Mass @ Rvir + Stellar Mass @ .2Rvir)')
    plt.ylabel('Gas Mass (Gas Mass @ Rvir)')
    plt.xscale('log')
    plt.yscale('log')
    plt.xlim(10**(7.8), 10**(14))
    plt.ylim(10**(-2), 10**(14))
    plt.plot(x, y2)
    plt.scatter(halo_mass_list, gas_mass_list, c=SSFR_list, s = 1, norm=matplotlib.colors.LogNorm())
    cbar= plt.colorbar()
    cbar.set_label('SSFR in 1/yr')
    plt.legend(('Omega Baryon / Omega Matter = 0.167',  'Host Galaxies'))
    
    #plt.subplot(523)
    #plt.title('Scale Factor %s, Stellar Mass Ratio vs Halo Mass' % str(a/1000))
    #plt.xlabel('Halo Mass (Gas Mass @ Rvir + Dark Matter Mass @ Rvir + Stellar Mass @ .2Rvir)')
    #plt.ylabel('Stellar Mass / Halo Mass')
    #plt.xscale('log')
    #plt.yscale('log')
    #plt.xlim(10**(7.8), 10**(14))
    #plt.ylim(1.0, 10**-6)
    #plt.plot(x, y1)
    #plt.scatter(halo_mass_list, stellar_ratio_list, c=SFR_list, s = 1, cmap='RdBu')
    #cbar= plt.colorbar()
    #cbar.set_label('SFR in MSun/yr')
    #plt.scatter(satellite_all_mass_mvir, gas_ratio_satellite, s = 1, c = 'r')
    #plt.legend(('Omega Baryon / Omega Matter = 0.167',  'Host Galaxies'))
    
    plt.subplot(222)
    plt.title('Stellar Mass vs Halo Mass at Redshift %s' % str((1/(a/1000)) - 1)[0:3])
    plt.xlabel('Halo Mass (Gas Mass @ Rvir + Dark Matter Mass @ Rvir + Stellar Mass @ .2Rvir)')
    plt.ylabel('Stellar Mass (Stellar Mass @ .2Rvir)')
    plt.xscale('log')
    plt.yscale('log')
    plt.xlim(10**(7.8), 10**(14))
    plt.ylim(10**(-2), 10**(14))
    plt.plot(x, y2)
    plt.scatter(halo_mass_list, stellar_mass_list, c=SSFR_list, s = 1, norm=matplotlib.colors.LogNorm())
    cbar= plt.colorbar()
    cbar.set_label('SSFR in 1/yr')
    plt.legend(('Omega Baryon / Omega Matter = 0.167',  'Host Galaxies'))
    
    #plt.subplot(525)
    #plt.title('Scale Factor %s, Baryon Mass Ratio vs Halo Mass' % str(a/1000))
    #plt.xlabel('Halo Mass (Gas Mass @ Rvir + Dark Matter Mass @ Rvir + Stellar Mass @ .2Rvir)')
    #plt.ylabel('Baryon Mass / Halo Mass')
    #plt.xscale('log')
    #plt.yscale('log')
    #plt.xlim(10**(7.8), 10**(14))
    #plt.ylim(0.0, 0.10)
    #plt.plot(x, y1)
    #plt.scatter(halo_mass_list, baryon_ratio_list, c=SFR_list, s = 1, cmap='RdBu')
    #cbar= plt.colorbar()
    #cbar.set_label('SFR in MSun/yr')
    #plt.legend(('Omega Baryon / Omega Matter = 0.167',  'Host Galaxies'))
    
    plt.subplot(223)
    plt.title('Baryon Mass vs Halo Mass at Redshift %s' % str((1/(a/1000)) - 1)[0:3])
    plt.xlabel('Halo Mass (Gas Mass @ Rvir + Dark Matter Mass @ Rvir + Stellar Mass @ .2Rvir)')
    plt.ylabel('Baryon Mass (Stellar Mass @.2Rvir + Gas @ Rvir)')
    plt.xscale('log')
    plt.yscale('log')
    plt.xlim(10**(7.8), 10**(14))
    plt.ylim(10**(-2), 10**(14))
    plt.plot(x, y2)
    plt.scatter(halo_mass_list, baryon_mass_list, c=SSFR_list, s = 1, norm=matplotlib.colors.LogNorm())
    cbar= plt.colorbar()
    cbar.set_label('SSFR in 1/yr')
    plt.legend(('Omega Baryon / Omega Matter = 0.167',  'Host Galaxies'))
    
    #plt.subplot(527)
    #plt.title('Scale Factor %s, SFR vs Stellar Mass, ' % str(a/1000))
    #plt.xlabel('Stellar Mass @.2Rvir')
    #plt.ylabel('SFR in MSun/yr')
    #plt.xscale('log')
    #plt.yscale('log')
    #plt.xlim(10**2, 10**13)
    #plt.scatter(stellar_mass_list, SFR_list, s=1)
    
    
    plt.subplot(224)
    plt.title('SSFR vs Stellar Mass at Redshift %s' % str((1/(a/1000)) - 1)[0:3])
    plt.xlabel('Stellar Mass @ .2Rvir')
    plt.ylabel('SSFR in 1/yr')
    plt.xscale('log')
    plt.yscale('log')
    plt.xlim(10**2, 10**13)
    plt.ylim(10**-13, 10**-7)
    plt.scatter(stellar_mass_list, SSFR_list, s=1)
    plt.legend(('Host Galaxies', 'blank'))

    plt.savefig('%s/thesis_baryon_SFR%s.png' % (output_dir, scale), bbox_inches='tight')
    plt.close()
```
