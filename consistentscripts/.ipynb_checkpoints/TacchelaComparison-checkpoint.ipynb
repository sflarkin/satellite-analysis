```python
import yt
import glob
import numpy as np
import argparse
import itertools
import matplotlib.pyplot as plt
from satellite_analysis.catalogreaders import consistentcatalogreader as consistent
from satellite_analysis.catalogreaders import tomercatalogreader as tomer

#for future args
input_dir = '/Users/user1/documents/GEN3/stellarmassrelation'
out_dir = '/Users/user1/documents/GEN3/stellarmassrelation/Nircomparison'

#Step 1: read and store all of the data for the rockstar results

#Step 1a: find all of the scale factors and VELA sim numbers for all of the simulations
catalog_files = glob.glob('%s/VELA*/*' % input_dir)
#print(catalog_files)
catalog_scales = []
catalog_numbers = []
for file in catalog_files:
    catalog_scales.append(file[-9:-6])
    catalog_numbers.append(file[-20:-18])
catalog_scales = np.unique(catalog_scales)
catalog_scales.sort()
catalog_numbers = np.unique(catalog_numbers)
catalog_numbers.sort()

#Step 1b: find all of the VELA directories, and set up the lists to write the data to 
values_rstar = [[] for x in range(len(catalog_scales))]

for number, scale in itertools.product(catalog_numbers, catalog_scales):
    try:
        readfile = open('%s/VELA%s/halomass%s.ascii' % (input_dir, number, scale))
        lines = readfile.readlines()
        halomass_lines = []
        for line in lines:
            halomass_lines.append(line.split())
        del halomass_lines[0]
        #find the largest halo
        count = 0
        largestmass = 0
        largestindex = 0
        while count < (len(halomass_lines)):
            if float(halomass_lines[count][2]) > largestmass:
                largestmass = float(halomass_lines[count][2])
                largestindex = count
            count = count + 1
        scale_index = [pos for pos, char in enumerate(catalog_scales) if char == scale][0]
        #now extract the data needed for comparison table
        #the data is stored in a tuple of form (VELA Sim Number, Scale Factor, Mvir, )
        values_rstar[scale_index].append((number, scale, float(halomass_lines[largestindex][2])*.7/10**12,
                                          float(halomass_lines[largestindex][4])*.7/10**10, 
                                          float(halomass_lines[largestindex][6])*.7/10**10, 
                                          float(halomass_lines[largestindex][8])*.7/10**10))
        
    except FileNotFoundError:
            print('No file: %s/VELA%s/halomass%s.ascii' % (input_dir, number, scale))

#Step 2: now find the same data for Nir's catalog

#set up the lists to add the values to 
values_Nir = [[] for x in range(len(catalog_scales))]

for number in catalog_numbers:
    tomer.read_tomer(number)
    tomer_scales = [str(scale)[2:5] for scale in tomer.tomer_list['aexpn'].tolist()]
    tomer_010rvir = tomer.tomer_list['Mstars(0.1rvir)'].tolist()
    tomer_015rvir = tomer.tomer_list['Mstar(0.15rvir)[Msun]'].tolist()
    tomer_020rvir = tomer.tomer_list['Mstars(0.2rvir)'].tolist()
    tomer_Mvir = tomer.tomer_list['M_rvir[Msun]'].tolist()
    for tomer_scale_index in range(len(tomer_scales)):
        scale_index = [pos for pos, char in enumerate(catalog_scales) if char == tomer_scales[tomer_scale_index]][0]
        values_Nir[scale_index].append((number, tomer_scales[tomer_scale_index],
                                        float(tomer_Mvir[tomer_scale_index])/10**12,
                                        float(tomer_010rvir[tomer_scale_index])/10**10,
                                        float(tomer_015rvir[tomer_scale_index])/10**10,
                                        float(tomer_020rvir[tomer_scale_index])/10**10))

        
#now make a table for each snapshot

#for scale in catalog_scales:
```

```python
#now we go about making a table for all of the data for each snapshot for all sims with a snapshot of that scale factor
#to do this, we start with a loop over all of the scale factors from the consistent catalogs
for index in range(len(catalog_scales)):
    #inside this loop, we find the number of VELA's with a snap of the desired scale factor, and loop over each
    #to make the data that will make up the row of an individual column
    
    #find all of the VELA numbers for each index of tomer's catalog of this scale factor to compare inside the next
    #loop over all of the simulations rockstar found data for at this snapshot
    tomer_VELA_numbers = [values_Nir[index][x][0] for x in range(len(values_Nir[index]))]
    
    #we also make the list for the data from each individual simulation to be stored to for final table construction
    table_data = []
    
    for number_sims in range(len(values_rstar[index])):
        #from here, we find the VELA number of this snap
        VELA_number = values_rstar[index][number_sims][0]
        rstar_Mvir = values_rstar[index][number_sims][2]
        rstar_010rstar = values_rstar[index][number_sims][3]
        rstar_015rstar = values_rstar[index][number_sims][4]
        rstar_020rstar = values_rstar[index][number_sims][5]
        #now check to see if there is a corresponding value from tomer's catalog
        if VELA_number in tomer_VELA_numbers:
            #find the list index that matches the VELA number of tomer's catalog
            list_index = [pos for pos, char in enumerate(tomer_VELA_numbers) if char == VELA_number][0]
            Nir_Mvir = values_Nir[index][list_index][2]
            Nir_010rstar = values_Nir[index][list_index][3]
            Nir_015rstar = values_Nir[index][list_index][4]
            Nir_020rstar = values_Nir[index][list_index][5]
        #if there isnt, make all of the values 'N/A', to mark that there is no data for these
        else:
            Nir_Mvir = 'N/A'
            Nir_010rstar = 'N/A'
            Nir_015rstar = 'N/A'
            Nir_020rstar = 'N/A'
            
        #now, we add the data for the table_data
        table_data.append([VELA_number, rstar_Mvir, Nir_Mvir, rstar_010rstar, Nir_010rstar, rstar_015rstar,
                          Nir_015rstar, rstar_020rstar, Nir_020rstar])
        
        #now plot the table and save it
        fig = plt.figure(figsize=(30, 10))
        ax = fig.add_subplot(111)
        clust_data = np.random.random((10,3))
        columnlabels = ('VELA Number', 'Rockstar Mvir', 'Nir Mvir', 'Rockstar Stellar Mass .1Mvir(10^12 Msun)', 
                        'Nir Stellar Mass .1Mvir(10^12 Msun)', 'Rockstar Stellar Mass .15Mvir(10^10 Msun)',
                        'Nir Stellar Mass .15Mvir(10^10 Msun)', 'Rockstar Stellar Mass .2Mvir(10^10 Msun)',
                        'Nir Stellar Mass .2Mvir(10^10 Msun)')
        ax.axis('tight')
        ax.axis('off')
        ax.table(cellText=table_data,colLabels=columnlabels,loc='top')
        plt.savefig('{}/table{}.png'.format(out_dir, catalog_scales[index]), bbox_inches='tight')
        plt.close()
        
```

```python
input_dirs = glob.glob('/Users/user1/documents/GEN3/VELA*/hlists')
input_dirs.sort()

VELA_numbers = ['06', '07', '08', '09', '10', '11', '12', '13', '14', '15']

for simulation_index in range(len(input_dirs)):
    print('Finding Values for directory', input_dirs[simulation_index])
    print('Using VELA number', VELA_numbers[simulation_index])
    #first get the tomer data
    tomer.read_tomer(VELA_numbers[simulation_index])
    tomer_scales = [str(scale)[2:5] for scale in tomer.tomer_list['aexpn'].tolist()]
    r_vir_tomer = tomer.tomer_list['r_vir[kpc]'].tolist()
    
    #now get the rockstar data
    consistent.consistent_catalog_reader(input_dirs[simulation_index], subhalos='True')
    
    #now loop over each consistent catalog entry, checking to see if there is a corresponding tomer scale for each
    #consistent file index
    for index in consistent.snapshot_index:
        scale_check = consistent.consistent_file_index[index]
        tomer_number = [pos for pos, number in enumerate(tomer_scales) if number == scale_check]
        if tomer_number == [] or len(tomer_number) > 1:
            print('Could not find corresponding Tomer Snapshot Data for snap', scale_check)
        else:
            print('Success')
            tomer_rvir = r_vir_tomer[tomer_number[0]]
            consistent_rvir = float(consistent.halo_data_largest[index][0][0][11])
            print('tomer', tomer_rvir)
            print('consistent', consistent_rvir)
            print(float(scale_check)/1000)
            print(tomer_rvir, consistent_rvir*(float(scale_check)/1000)/1.43)
```
